```python
import pandas as pd
import yaml
from pathlib import Path
import os
```

```python
candidate_paths = Path.cwd().parent.parent / "dataset.yaml",

config_path = next((path for path in candidate_paths if path.exists()), None)
if config_path is None:
    raise FileNotFoundError("Could not find dataset.yaml from the current working directory")

with config_path.open("r") as file:
    dataset_paths = yaml.safe_load(file)

print(dataset_paths)
```

```python
train_path = dataset_paths["TITANIC_TRAIN"]
test_path = dataset_paths["TITANIC_TEST"]

base_dir = '../../'

train_data = pd.read_csv(os.path.join(base_dir, train_path))
test_data = pd.read_csv(os.path.join(base_dir, test_path))
```

```python
train_data.head()
```

```python
train_data.shape
```

```python
train_data.isna().sum()
```

```python
train_data['Embarked'].value_counts()
train_data.fillna({'Embarked':train_data['Embarked'].mode()[0]}, inplace=True)
train_data['Cabin'].value_counts()
train_data.fillna({'Cabin':train_data['Cabin'].mode()[0]}, inplace=True)
train_data['Age'].mean()
train_data.fillna({'Age':round(train_data['Age'].mean(), 2)}, inplace=True)
train_data.isna().sum()
```

```python
train_data.head()
```

```python
train_data.drop(columns=['PassengerId','Name', 'Ticket', 'Cabin'], inplace=True)
train_data.shape
```

```python
data = pd.get_dummies(data=train_data, columns=['Pclass','Sex','SibSp','Parch','Embarked'], drop_first=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
data['Age_scaled'] = scaler.fit_transform(data[['Age']])
data['Fare_scaled'] = scaler.fit_transform(data[['Fare']])
```

```python
data.drop(columns=['Age','Fare'], inplace=True)
```
